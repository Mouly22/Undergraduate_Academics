```python
class myList:
    def __init__(self, *args):
        self.lst = []
        for x in args:
            self.lst.append(x)
            
    def sum(self):
        self.summ = 0
        for m in self.lst:
            self.summ += int(m)
        print(f'Sum: {self.summ}')
            
    def merge(self, *args2):
        for g in args2:
            self.lst.append(g)
            
    def average(self):
        if len(self.lst) == 0:
            avg = 0
        else:
            avg = self.summ / len(self.lst)
        print(f'Average: {avg}')
        


l1 =  myList(2,3,4,5,6)
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print('-----------------------------')
l2 =  myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```
