#  Question-1
**Implement the design of the Motorcycle class so that the following output is produced:**

**<font color="Blue">[Your code should work for any number of services added in the add_Services method]</font>**


**Output:**\
No. of Motorcycle = 0\
\=========================\
Name: Yamaha MT-15\
Year of manufacture: 2019\
Services: Oil change, Replace coolant\
\=========================\
Name: Suzuki V-Strom 650\
Year of manufacture: 2020\
Services: Rear suspension check, New air filter, Oil change\
\=========================\
Name: Honda Gold Wing\
Year of manufacture: 2017\
Services: Adjustment of throttle and clutch, New oil filter\
\=========================\
No. of Motorcycle = 3


```python
# Write your code here

class Motorcycle:
    count = 0
    def __init__(self, name, year):
        self.name = name
        self.year = year
        self.lst = []
        
    def add_Services(self, *args):
        for i in args:
            self.lst.append(i)
            
    def printMotorcycleDetail(self):
        print(f'Name:{self.name}')
        print(f'Year of manufacture:{self.year}')
        var = "Services:"
        for elm in self.lst:
            var += elm + ', '
        var = var.rstrip(', ')
        print(var)
        
#=================================================
# Do not change the following lines of code.
print('No. of Motorcycle =', Motorcycle.count)
m1 = Motorcycle('Yamaha MT-15', 2019)
m1.add_Services('Oil change', 'Replace coolant')
m2 = Motorcycle('Suzuki V-Strom 650', 2020)
m2.add_Services('Rear suspension check', 'New air filter', 'Oil change')
m3 = Motorcycle('Honda Gold Wing', 2017)
m3.add_Services('Adjustment of throttle and clutch', 'New oil filter')
print("=========================")
m1.printMotorcycleDetail()
print("=========================")
m2.printMotorcycleDetail()
print("=========================")
m3.printMotorcycleDetail()
print("=========================")
print('No. of Motorcycle =', Motorcycle.count)
```

#  Question-2
Implement the design of the Traveller​ ​class so that the following output\
is produced:

**<font color="blue">[Your code should work for any number of tickets added in the buyTickets method]</font>**


**Output:**\
Total Traveller: 0\
\=========================\
Name: Naruto\
Tickets: Mist Village, Rain Village\
Cost: 15000\
\=========================\
Name: Minato\
Tickets: Cloud Village, Sand Village, Stone Village\
Cost: 21200\
\=========================\
Name: Kushina\
Tickets: Leaves Village\
Cost: 3000\
\=========================\
Total Traveller: 3


```python
# Write your codes here.

class Traveller:
    count = 0
    def __init__(self, name):
        self.name = name
        self.lst = []
        self.lst2 = []
        self.ticket = 'Tickets:'
        self.cost = 0
        
        
    def buyTickets(self, *args):
        for i in args:
            self.lst.append(i)
            
        for x in range(len(self.lst)//2):
            self.lst2.append([self.lst[0 + 2* x], self.lst[1 + 2* x]])
            
        for j in self.lst2:
            self.ticket += j[0] + ', '
            self.cost += j[1]
        self.ticket = self.ticket.rstrip(', ')
            
        
        
    def printDetail(self):
        print(f'Name: {self.name}')
        print(self.ticket)
        print(f'Cost:{self.cost}')
             

#=======================================================
# Do not change the following lines of code.
print("Total Traveller:", Traveller.count)
t1 = Traveller("Naruto")
t1.buyTickets("Mist Village",5000,"Rain Village", 10000)
t2 = Traveller("Minato")
t2.buyTickets("Cloud Village",15000,"Sand Village",1200,"Stone Village", 5000)
t3 = Traveller("Kushina")
t3.buyTickets("Leaves Village", 3000)
print("=========================")
t1.printDetail()
print("=========================")
t2.printDetail()
print("=========================")
t3.printDetail()
print("=========================")
print("Total Traveller:", Traveller.count)
```

```python
# Write your codes here.
class Pasta:
    pasta_plate_count = 0
    
    def __init__(self, p_type):
        self.p_type = p_type
    def set_ingredients(self, shape, cal, fat, protein, carbohydrate):
        self.shape = shape
        self.cal = cal
        self.fat = fat
        self.protein = protein
        self.carbohydrate = carbohydrate
        
    def display_all_info(self):
        print(f'Pasta Type: {self.p_type}')
        print(f'Shape: {self.shape}')
        print(f'Cal: {self.cal}')
        print(f'Fat:{self.fat}')
        print(f'Protein: {self.protein}')
        print(f'Carbohydrate:{self.carbohydrate}')
    


#=======================================================
# Do not change the following lines of code.
print("Pasta Plate Count:", Pasta.pasta_plate_count)
print("=======================")
s1 = Pasta("Chicken")
s1.set_ingredients("Farfalle",250, 11, 40, 10)
s1.display_all_info()
print("------------------------------------")
s2 = Pasta("Beef")
s2.set_ingredients("Pipe Rigate",150, 10.5, 10, 10)
s2.display_all_info()
print("------------------------------------")
s3 = Pasta("Mushrooms")
s3.set_ingredients("Spaghetti",500, 50, 20, 10)
s3.display_all_info()
print("=======================")
print("Pasta Plate Count:", Pasta.pasta_plate_count)
```

Year of release: 2021\
Features: 512GB PCIe NVMe SSD, Intel Core i7-1165G7 CPU\

```python
# Write your code here

class Laptop:
    count = 0
    def __init__(self, name, release):
        self.name = name
        self.release = release
        self.lst = []
        
    def add_Features(self, *args):
        for elm in args:
            self.lst.append(elm)
            
    def printLaptopDetail(self):
        print(f'Model Name: {self.name}')
        print(f'Year of release: {self.release}')
        var = 'Features:'
        for i in self.lst:
            var += i + ', '
        var = var.rstrip(', ')
        print(var)
        

#========================================================================
# Do not change the following lines of code.

print("No. of Laptop =", Laptop.count)
l1 = Laptop('Macbook Pro', 2021)
l1.add_Features('Apple M1 Chip' , '16-core Neural Engine', '8-core GPU')
l2 = Laptop('Razer Blade Pro 17', 2020)
l2.add_Features('Intel Core i7-10875H CPU' , 'Nvidia RTX 3080 16GB GPU')
l3 = Laptop('HP SPECTRE X360 14', 2021)
l3.add_Features('512GB PCIe NVMe SSD' ,  'Intel Core i7-1165G7 CPU')
print("=========================")
l1.printLaptopDetail()
print("=========================")
l2.printLaptopDetail()
print("=========================")
l3.printLaptopDetail()
print("=========================")
print('No. of Laptop =',Laptop.count)
```

#  Question-6
Implement the design of the DemonSlayer class so that the following output is produced:

**Output:**\
No of demon slayers: 0\
\=========================\
Name: Tanjiro\
Age: 15\
Style: Water Breathing\
Total Kill: 120\
\=========================\
Name: Rengoku\
Age: 20\
Style: Flame Breathing\
Total Kill: 950\
\=========================\
Name: Zenitsu\
Age: 16\
Style: Thunder Breathing\
Total Kill: 80\
\=========================\
No of demon slayers: 3

```python
# Write your code here
class DemonSlayer:
    count = 0
    
    def __init__(self, name, age, style, tkill):
        self.name = name
        self.age = age
        self.style = style
        self.tkill = tkill
    def addKill(self, tkill):
        self.tkill += tkill
        
    def dismissKill(self, bkill):
        self.bkill = bkill
        if self.tkill >= self.bkill:
            self.tkill -= bkill
        else:
            self.bkill = self.tkill
        
    def printDetails(self):
        print(f'Name:{self.name}')
        print(f'Age:{self.age}')
        print(f'Style:{self.style}')
        print(f'Total Kill: {self.tkill}')


#========================================================================
# Do not change the following lines of code.

print('No of demon slayers:', DemonSlayer.count)
print("=========================")
d1 = DemonSlayer("Tanjiro", 15, "Water Breathing", 100)
d1.addKill(20)
d1.printDetails()
print("=========================")
d2 = DemonSlayer("Rengoku", 20, "Flame Breathing", 1000)
d2.dismissKill(50)
d2.printDetails()
print("=========================")
d3 = DemonSlayer("Zenitsu", 16, "Thunder Breathing", 80)
d3.dismissKill(100)
d3.printDetails()
print("=========================")
print('No of demon slayers:', DemonSlayer.count)
```
